```python
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
bank_0 = 1000
odds = 0.70
payout_factor = 0.8
risk_pc = 0.05
gamenights = 100
iters = 2000
```

```python
progress = []
for k in range(iters):
    progress.append([])
for j in range(iters):
    bank = bank_0
    for i in range(gamenights):
#         if bank <= 0: break
        win = random.uniform(0, 1) < odds
        if win:
            result = bank * risk_pc * payout_factor
        else:
            result = -1 * bank * risk_pc
        bank += result
        progress[j].append(bank)
```

```python
plt.figure(1, figsize=(12,12))
for j in range(iters):
#     plt.scatter(range(gamenights), progress[j], alpha=0.002, c='black', s=100)
    plt.plot(progress[j], alpha=0.002, linewidth=15, c='black')
plt.xlim([0, gamenights])
plt.ylim([0, 12000])
appreciation_factor = odds*(1 + payout_factor * risk_pc) + (1-odds) * (1-risk_pc)
plt.plot([bank_0 * appreciation_factor ** i for i in range(gamenights)], linewidth=7, alpha = 0.5, c='cyan')
```
