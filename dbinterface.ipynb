```python
def get_players_by_date(date, games=games):
    q = {'date': date}
    players = [g['box'][0]['players'] + g['box'][1]['players'] for g in games.find(q)]
    players = reduce(lambda a,b: a+b, players)
    return players
```

```python
get_players_by_date(date1)[:2]
```

```python
def convert_to_stats(player):
    stats = {}
    stats['assists'] = player['ast']
    stats['blocks'] = player['blk']
    stats['steals'] = player['stl']
    stats['rebounds'] = player['trb']
    stats['turnovers'] = player['tov']
    stats['points'] = player['ft'] + 2 * player['fg'] + 3 * player['fg3']
    stats['made_3s'] = player['fg3']
    return stats
```

```python
sorted([convert_to_stats(p) for p in get_players_by_date(date1)],key=lambda x: x['points'],reverse=True)
```
